```python
# to install module in JupyterLite
import piplite
await piplite.install("folium")
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
#Define coordinates of where we want to center our map
Penn_Campus_coords = [39.952305, -75.193703]

#Create the map
my_map = folium.Map(location = Penn_Campus_coords, zoom_start = 15)

#Display the map
my_map
```

```python
#Define the coordinates we want our markers to be at
philly_coords = [39.952583, -75.165222]
Penn_Campus_coords = [39.952305, -75.193703]
Van_Pelt_coords = [39.9527, -75.1936]

#Add markers to the map
folium.Marker(philly_coords, popup = 'Philadelphia').add_to(my_map)
folium.Marker(Penn_Campus_coords, popup = 'University of Pennsylvania').add_to(my_map)
folium.Marker(Van_Pelt_coords, popup = 'Van Pelt Library').add_to(my_map)

#Display the map
my_map
```

```python
#Using polygon markers with colors instead of default markers
polygon_map = folium.Map(location = Penn_Campus_coords, zoom_start = 16)

philly_coords = [39.952583, -75.165222]
Penn_Campus_coords = [39.952305, -75.193703]
Van_Pelt_coords = [39.9527, -75.1936]

#Add markers to the map
folium.RegularPolygonMarker(philly_coords, popup = 'Philadelphia', fill_color = '#00ff40',
                            number_of_sides = 3, radius = 7).add_to(polygon_map)
folium.RegularPolygonMarker(Penn_Campus_coords, popup = 'University of Pennsylvania', fill_color = '#bf00ff',
                            number_of_sides = 5, radius = 7).add_to(polygon_map)
folium.RegularPolygonMarker(Van_Pelt_coords, popup = 'Van Pelt Library', fill_color = '#ff0000',
                           number_of_sides = 8, radius = 7).add_to(polygon_map)

#Display the map
polygon_map
```
